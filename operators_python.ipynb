# operators in python

## 6 types of operatorsin python
1. arithematic opperators
2. assignment opperators
3. comparision opperators
4. logical opperators
5. membership opperators
6. identity opperators

```python
# arithematic opperators - + ,-, *, /, //, %
a=10
b=20
c=20
d=30
print(a+b)
print(b-c)
print(b*d)
print(d/a)
print(d//a)
print(d%a)                
```

```python
# assingment opperators - +=, -=, *=, /=, //=, %=
a=100
a+=50
print(a)
```

```python
b=200
b-=400
print(b)
```

```python
c=50
c*=100
print(c)
```

```python
# comparision opperators - >,<,>=,<=,==,!=
a=10
b=20
c=20
d=30
print(a>b)
print(b==c)
print(b!=c)
print(b>=a)
print(b>=c)
print(b<=a)
```

```python
# logical opperator - and, or, not
"""
and - both conditions are true then results is true , rest case it false
or - either one condition is true the result is true , rest case is false
and truth table
T T = T
T F = F
F T = F
F F = F
or truth table
T T = T
T F = T
F T = T
F F = F
"""

a=10
b=20
c=20
d=30
print(a>b and b==c)
print(a<b and c==b)
print(a>b or b==c)
print(not a>b and b==c)
```

```python
# membership opperator - in, not in
# volidating membership are present in that sequence or group
names=["vikas", "shivaji", "sidh", "chandra"]
print("vikas" in names)
print("shivaji"not in names)
print("Vikas"not in names)
print("vikram"in names)
```

```python
# identity opperators - is, is not
a=10000
b=10000
print(a is b)
print(id(a))
print(id(b))
```
