````python
"""
write a program to find the student grades based on marks
1. if student marks in between 100 to 80 then grade is A
2. if student marks in between 79 to 60 then grade is B
3. if student marks in between 59 to 40 then grade is C
4. if student marks in between 39 to 25 then grade is D
5. if student marks in between 24 to 0 then grade is Fail
6. if other than above marks expected output should be plese enter correct marks!
"""
marks=int(input("enter student marks here:"))
if (marks>=80 and marks<=100):
    print(marks,"is A grade")
elif (marks>=60 and marks<=79):
    print(marks,"is B grade")
elif (marks>=40 and marks<=59):
    print(marks,"is C grade")
elif (marks>=25 and marks<=39):
    print(marks,"is D grade")
elif (marks>=0 and marks<=24):
    print(marks,"is Fail")
else:
    print(marks,"is not correct marks, plese enter correct marks")
````

````python
# write a program to find the largest number from three numbers
if num1 >= num2 and num1 >= num3:
        largest = num1
        print(f"The largest number is: {largest} (from the 'if' block)")
elif num2 >= num1 and num2 >= num3:
        largest = num2
        print(f"The largest number is: {largest} (from the 'elif' block)"
else:
        largest = num3
        print(f"The largest number is: {largest} (from the 'else' block)")
    
    return largest
````

````python
# write a program to find the leap year or non leap year
if year % 400 == 0:
        print(f"{year} is a LEAP YEAR (Divisible by 400).")
elif year % 100 == 0:
        print(f"{year} is NOT a leap year (Divisible by 100, but not 400).")
elif year % 4 == 0:
        print(f"{year} is a LEAP YEAR (Divisible by 4, but not 100).")
else:
        print(f"{year} is NOT a leap year (Not divisible by 4).")
````

````python
# write a program to check login credentials using conditional statements

# Define the correct (secret) credentials
CORRECT_USERNAME = "student_user"
CORRECT_PASSWORD = "Pa$$word123"

def check_login():
    """
    Prompts the user for a username and password and verifies them.
    """
    print("--- Login System ---")
    
    # Get user input for the attempt
    user_input = input("Enter username: ")
    pass_input = input("Enter password: ")

    # Condition 1: Check for successful login (both credentials match)
    if user_input == CORRECT_USERNAME and pass_input == CORRECT_PASSWORD:
        print("\nSUCCESS! Welcome to the system, student_user!")
        print("You are logged in.")

    # Condition 2: If the password is correct but the username is wrong, 
    # it means the user made a mistake in the username.
    # Note: For security, real systems usually just say "Invalid credentials."
    # but for this exercise, we show the specific failure paths.
    elif user_input != CORRECT_USERNAME:
        print("\nACCESS DENIED. Incorrect username. Please try again.")

    # Condition 3 (The 'else' block): If we reach here, the username MUST be correct 
    # (because it didn't hit the elif block), but the full 'if' statement failed, 
    # meaning the password must be wrong.
    else: 
        print("\nACCESS DENIED. Incorrect password. Please try again.")
````
